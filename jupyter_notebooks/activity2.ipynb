```python
# importing required packages
import pandas as pd
import requests
import hashlib
import json
import string
```

```python
starting_characters = list('3')+list(string.ascii_uppercase)
#starting_characters = list(string.digits)+list(string.ascii_uppercase)
```

```python
# APPROACH 1 - using nameStartsWith parameter, not offset
public_key = '486e33cb6586570842beb1fdc21db6c2'
private_key = '4893102e65894f0350a76693dff1b2bc9418a52a'
ts = 1
hash = hashlib.md5((str(ts)+private_key+public_key).encode()).hexdigest()
address = 'http://gateway.marvel.com/v1/public/characters'
df = pd.DataFrame()

for i in starting_characters:
    parameters = {
        'apikey':public_key,
        'ts':ts,
        'hash':hash,
        'limit':100,
        'nameStartsWith':i
    }
    response = requests.get(address, params=parameters)
    results = response.json()
    df_new = pd.json_normalize(results['data']['results'])
    df = pd.concat([df, df_new], ignore_index=True)

character_df = df[['id','name','events.available','series.available','stories.available','comics.available']]
character_df
```

```python
#checking the list of unique starting characters in character_df.name
character_df.name.str[:1].unique()
```

```python
# APPROACH 2 - using only offset, not nameStartsWith parameter
public_key = '486e33cb6586570842beb1fdc21db6c2'
private_key = '4893102e65894f0350a76693dff1b2bc9418a52a'
ts = 1
hash = hashlib.md5((str(ts)+private_key+public_key).encode()).hexdigest()
address = 'http://gateway.marvel.com/v1/public/characters'
df = pd.DataFrame()

for i in range(0,16):
    parameters = {
        'apikey':public_key,
        'ts':ts,
        'hash':hash,
        'limit':100,
        'offset':100*i
    }
    response = requests.get(address, params=parameters)
    results = response.json()
    df_new = pd.json_normalize(results['data']['results'])
    df = pd.concat([df, df_new], ignore_index=True)

character_df = df[['id','name','events.available','series.available','stories.available','comics.available']]
character_df
```

```python
#checking the list of unique starting characters in character_df.name
character_df.name.str[:1].unique()
```
